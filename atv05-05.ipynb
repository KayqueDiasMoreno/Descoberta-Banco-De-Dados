```python
from pymongo import MongoClient
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://otaboanense.com.br/'
```

```python
req = requests.get(url)
html = req.text
```

```python
soup = BeautifulSoup(html, 'html.parser')   
```

```python
items = soup.find_all('div', class_= 'pt-4')
```

```python
for item in items:
    print(item.find('h3').text)
```

```python
manchetes_site = []

for item in soup.find_all('div', class_= 'pt-4'):
    manchetes = {
        'Manchete:' : item.find('h3').text
    }
    manchetes_site.append(manchetes)

display(manchetes_site)
```

```python
client = MongoClient('mongodb://root:rootpass@localhost:27017')
client.server_info()['version']
```

```python
db = client['noticiario']
collection = db['manchetes']
```

```python
collection.insert_many(manchetes_site)
```

```python
for doc in collection.find():
    print(doc)
```
